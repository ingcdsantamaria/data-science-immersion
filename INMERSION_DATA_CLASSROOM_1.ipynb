importaremos herramientas de python para ver mejor ese tipo de graficos

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))


grafica = sns.histplot(data=inmuebles, x = 'Precio_Millon', kde = True, hue='Tipo')

grafica.set_title('Distribucion de valores de los inmuebles en bogota')

plt.xlim((50,1000))


plt.savefig('/content/drive/MyDrive/inmersion data science/valor_inmuebles.png',format='png')

plt.show()
```

**DESAFIO**



1.   Estudiar mejor el histograma de valores seleccionar 3 tipos de inmuebles(refinar el grafico: Titulos, aumentar el tamaño de labels, conclusion de la informacion) 




```python
plt.figure(figsize=(10,6))


tipos_seleccionados = ['Local', 'Bodega', 'Lote']
inmuebles_seleccionados = inmuebles[inmuebles['Tipo'].isin(tipos_seleccionados)]

grafica = sns.histplot(data=inmuebles_seleccionados, x='Precio_Millon', kde=True, hue='Tipo')
grafica.set_title('Distribucion de valores de 3 inmuebles', fontsize=18)
grafica.set_xlabel('Precio (en millones)', fontsize=16)
grafica.set_ylabel('Cantidad', fontsize=16)

plt.xlim((50,6000))

plt.savefig('/content/drive/MyDrive/inmersion data science/valor_3_inmuebles.png',format='png')
plt.show()
```

**CONCLUSIONES:**

1.   Basandonos en la altura de las barras de color azul(Locales), esta claro que la cantidad de locales en venta es mayor a la de bodegas y lotes sin embargo evidenciamos que los que se encuentran con un valor entre 100 millones a 600 millones quien supera la cantidad de inmuebles en venta son las bodegas.
2.   Basandonos en la altura de las barras de colo verde(Lote), esta claro que tiene las ventas mas bajas y los valores mas economicos esto con el fin de visualizar posibles compras



2.   Precio del m2 por barrio y investigar que grafica se adecue mas para esta nueva variable y hacer el grafico mas adecuado para esta nueva variable(precio del metro cuadrado es igual a precio/m2=precioMetroCuadrado) 

```python
inmuebles['valor_metroCuadrado'] = inmuebles.Precio_Millon / inmuebles.Area

plt.figure(figsize=(10,6))

grafica = sns.scatterplot(data=inmuebles, x = 'valor_metroCuadrado',y = 'Precio_Millon', hue='Tipo')

grafica.set_title('Distribucion de valor de metro cuadrado por valor total', fontsize=18)
grafica.set_xlabel('Valor por metro cuadrado (en millones)', fontsize=16)
grafica.set_ylabel('Valor total', fontsize=16)

plt.savefig('/content/drive/MyDrive/inmersion data science/valor_metro_cuadrado.png',format='png')
plt.xlim((5,95))

plt.show()
```

**Conclusiones:** en el grafico de dispercion nos sirve para ver los inmuebles sobrevaloradas o subvaloradas en relación con su valor por metro cuadrado y valor total
